```
import pandas as pd

data = "/content/drive/MyDrive/STUDY 1_2_3 FULL DATASET - Primary Model.csv"
df = pd.read_csv(data)

df.head()
```

```
#creating a dataframe with all training data except the target column
X = df.drop(columns=['TARGET good_responder_psa_3m_2ymean_>0.2psaincrease'])

X.head()
```

```
#separating target values
y = df['TARGET good_responder_psa_3m_2ymean_>0.2psaincrease'].values

#view target values
y[0:5]
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1, stratify=y)
```

```
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```
model = Sequential()
model.add(Dense(2024, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(100, activation='relu'))
model.add(Dense(60, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
model.fit(X_train, y_train, epochs=150, batch_size=32)
```

```
_, accuracy = model.evaluate(X_test, y_test)
print('Accuracy: %.2f' % (accuracy*100))
```
